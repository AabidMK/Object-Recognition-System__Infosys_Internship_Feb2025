```python
from ultralytics import YOLO
import cv2
import os

# Ensure correct path handling
model_path = os.path.join('run', 'detect', 'train3', 'weights', 'best.pt')

# Load the YOLOv8 model
model = YOLO(model_path)

# Open the video file
video_path = 'fruits.mp4'
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break  # Exit loop if video ends

    # Convert BGR (OpenCV format) to RGB (YOLO format)
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Run YOLOv8 inference
    results = model(frame_rgb)

    # Plot bounding boxes directly
    annotated_frame = results[0].plot()

    # Display the processed frame
    cv2.imshow('YOLOv8 Scaling Detection', annotated_frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
