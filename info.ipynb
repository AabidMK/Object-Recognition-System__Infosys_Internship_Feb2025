```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Count total images
image_files = [f for f in os.listdir(dataset_path) if f.endswith('.jpg')]
total_images = len(image_files)
print(f"Total number of images: {total_images}")

# Load annotation data (assuming it's in a structured format)
# Replace 'annotations.json' with your actual annotation file
import json
with open('/content/drive/MyDrive/coco2017_subset/annotation_subset/instances_train2017_subset.json', 'r') as f:
    annotations = json.load(f)

# Get class distribution
class_counts = {}  # Count images per class
# Update this based on your annotation format
for annotation in annotations['annotations']: # Iterate through the 'annotations' list in the JSON data
    category_id = annotation['category_id']  # Access the category ID

    # You might need to map category_id to category_name if needed
    # For example: categories_dict = {category['id']: category['name'] for category in annotations['categories']}
    # category_name = categories_dict[category_id]

    if category_id in class_counts:
        class_counts[category_id] += 1
    else:
        class_counts[category_id] = 1

# Plot class distribution
plt.figure(figsize=(12, 6))
plt.bar(class_counts.keys(), class_counts.values())
plt.title('Class Distribution')
plt.xlabel('Class ID')  # Changed to Class ID as we are using category_id
plt.ylabel('Number of Images')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Analyze image dimensions, aspect ratios, file sizes
widths, heights, aspect_ratios, file_sizes = [], [], [], []

for img_file in image_files[:100]:  # Limit to 100 images for speed
    img_path = os.path.join(dataset_path, img_file)
    # Get file size
    file_sizes.append(os.path.getsize(img_path) / 1024)  # Size in KB
    # Get image dimensions
    with Image.open(img_path) as img:
        width, height = img.size
        widths.append(width)
        heights.append(height)
        aspect_ratios.append(width / height)

# Plot distributions
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
axs[0, 0].hist(widths, bins=20)
axs[0, 0].set_title('Image Width Distribution')
axs[0, 1].hist(heights, bins=20)
axs[0, 1].set_title('Image Height Distribution')
axs[1, 0].hist(aspect_ratios, bins=20)
axs[1, 0].set_title('Aspect Ratio Distribution')
axs[1, 1].hist(file_sizes, bins=20)
axs[1, 1].set_title('File Size Distribution (KB)')
plt.tight_layout()
plt.show()

# Summary statistics
print(f"Width - Min: {min(widths)}, Max: {max(widths)}, Mean: {np.mean(widths):.2f}")
print(f"Height - Min: {min(heights)}, Max: {max(heights)}, Mean: {np.mean(heights):.2f}")
print(f"Aspect Ratio - Min: {min(aspect_ratios):.2f}, Max: {max(aspect_ratios):.2f}")
print(f"File Size (KB) - Min: {min(file_sizes):.2f}, Max: {max(file_sizes):.2f}, Mean: {np.mean(file_sizes):.2f}")
```

```python
# Analyze image brightness, contrast, color distribution
brightnesses = []
color_distributions = []  # For RGB images

for img_file in image_files[:100]:  # Limit to 100 images for speed
    img_path = os.path.join(dataset_path, img_file)
    with Image.open(img_path) as img:
        # Convert to numpy array
        img_array = np.array(img)
        # Calculate brightness (average pixel value)
        brightness = np.mean(img_array)
        brightnesses.append(brightness)

        # For color images, analyze RGB channels
        if len(img_array.shape) == 3 and img_array.shape[2] == 3:
            # Average values for R, G, B channels
            color_dist = [np.mean(img_array[:,:,0]), np.mean(img_array[:,:,1]), np.mean(img_array[:,:,2])]
            color_distributions.append(color_dist)

# Plot brightness distribution
plt.figure(figsize=(10, 6))
plt.hist(brightnesses, bins=20)
plt.title('Image Brightness Distribution')
plt.xlabel('Average Brightness')
plt.ylabel('Number of Images')
plt.show()

# Color distribution if we have color images
if color_distributions:
    r_vals = [c[0] for c in color_distributions]
    g_vals = [c[1] for c in color_distributions]
    b_vals = [c[2] for c in color_distributions]

    plt.figure(figsize=(10, 6))
    plt.hist(r_vals, bins=20, alpha=0.5, color='red', label='Red Channel')
    plt.hist(g_vals, bins=20, alpha=0.5, color='green', label='Green Channel')
    plt.hist(b_vals, bins=20, alpha=0.5, color='blue', label='Blue Channel')
    plt.title('RGB Channel Distributions')
    plt.xlabel('Average Channel Value')
    plt.ylabel('Number of Images')
    plt.legend()
    plt.show()
```

```python
import os
import cv2
import matplotlib.pyplot as plt
from pycocotools.coco import COCO

# Load COCO annotations
coco = COCO('/content/drive/MyDrive/coco2017_subset/annotation_subset/instances_val2017_subset.json')

# Get all image IDs
image_ids = coco.getImgIds()

def show_image_with_annotations(image_id, coco):
    image_info = coco.loadImgs(image_id)[0]

    # Correct file path handling
    image_path = os.path.join("/content/drive/MyDrive/coco2017_subset/test2017", image_info['file_name'])

    # Debugging step: Print the file path
    print("Loading image from:", image_path)

    # Load image
    image = cv2.imread(image_path)

    # Check if the image is loaded
    if image is None:
        print(f"Error: Unable to load image {image_path}")
        return

    # Convert BGR to RGB (OpenCV loads in BGR by default)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Get annotation IDs for the image
    annotation_ids = coco.getAnnIds(imgIds=image_id)
    annotations = coco.loadAnns(annotation_ids)

    # Draw bounding boxes
    for ann in annotations:
        x, y, w, h = ann["bbox"]
        cv2.rectangle(image, (int(x), int(y)), (int(x+w), int(y+h)), (255, 0, 0), 2)

    # Show image
    plt.figure(figsize=(8, 8))
    plt.imshow(image)
    plt.axis("off")  # Hide axis for better visualization
    plt.show()

# Show a sample image
show_image_with_annotations(image_ids[0], coco)
```
